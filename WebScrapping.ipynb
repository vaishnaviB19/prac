```python
response=requests.get(URL)
soup=bs4.BeautifulSoup(response.text,'html.parser')
```

```python
soup
```

```python
def getCustomer_name(soup):
    data_string=""
    customer_list=[]
    for item in soup.findAll("span", class_="a-profile-name"):
        data_string=data_string+item.get_text()
        customer_list.append(data_string)
        data_string=""
    return customer_list;
```

```python
customer_res = getCustomer_name(soup)
print("Customer names:",customer_res)
```

```python
def getReviews(soup):
    reviews_list=[]
    for item in soup.findAll("div", class_="a-expander-content reviewText review-text-content a-expander-partial-collapse-content"):
        reviews_list.append(item.get_text())
    return reviews_list;
```

```python
customer_rev = getReviews(soup)
print("Customer reviews:",customer_rev,"\n\n")
```

```python
def getRatings(soup):
    rating_list=[]
    for item in soup.findAll("i", class_="a-icon a-icon-star a-star-5 review-rating"):
        rating_list.append(item.get_text())
    return rating_list;
```

```python
customer_ratings = getRatings(soup)
print("Customer ratings:",customer_ratings,"\n\n")
```

```python
def getCommentTags(soup):
    tag_list=[]
    for item in soup.findAll("span", class_="a-size-base a-color-secondary"):
        tag_list.append(item.get_text())
    return tag_list;
```

```python
customer_tags = getCommentTags(soup)
print("Customer Tags:",customer_tags,"\n\n")
```
